```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import sys, os
%matplotlib inline
sys.path.append('../')
import coremagmodels as cm
```

```python
Nth = 100
Nph = Nth*2
dth = np.pi/Nth
colat_rad = np.linspace(dth/2, np.pi-dth/2, Nth)
lon_rad = np.linspace(dth/2, 2*np.pi-dth/2, Nph)
```

# Chaos-6

```python
mag = cm.coremagmodels.Chaos6()
T = 2010
l_max = 14
sv_lmax = 16
sa_lmax = 16
bSH = mag.get_shtcoeffs_at_t(T,l_max=l_max)
svSH = mag.get_SVshtcoeffs_at_t(T,l_max=sv_lmax)
saSH = mag.get_SAshtcoeffs_at_t(T,l_max=sa_lmax)
Br = mag.B_sht(bSH,Nth=Nth)
SV = mag.B_sht(svSH,Nth=Nth)
SA = mag.B_sht(saSH,Nth=Nth)

plt.figure(figsize=(14,6))
plt.subplot(131)
cm.plot.plot_base_DH(Br/1e6, title='Br, {}'.format(mag.name), newfig=False, clbl='mT')
plt.subplot(132)
cm.plot.plot_base_DH(SV/1e3, title='SV, {}'.format(mag.name), newfig=False, clbl='uT/yr')
plt.subplot(133)
cm.plot.plot_base_DH(SA/1e3, title='SA, {}'.format(mag.name), newfig=False, clbl='uT/yr^2')
```
